```python
data=data.dropna(subset=["Description","CustomerID"]) 
#since the comlumn named description and customer id is very import, if miss, just delete
```

```python
data.loc[:,"InvoiceDate"]=data.loc[:,"InvoiceDate"].astype("object") #change it to object 
#data[data["InvoiceDate"]]=data[data["InvoiceDate"]].astype(str)
```

```python
data.dtypes 
```

```python
#Get the year, month from InvoiceDate
data.loc[:,"invoicY&M"]=data.loc[:,"InvoiceDate"].dt.date
data.loc[:,"invoiceYear"]=data.loc[:,"InvoiceDate"].dt.year
data.loc[:,"invoiceMon"]=data.loc[:,"InvoiceDate"].dt.month
```
